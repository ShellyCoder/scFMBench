```python
import os
import torch
import pandas as pd
from DrugResponseEvaluator import DrugResponseEvaluator
```

```python
if __name__ == "__main__":
    # Set all paths to the example folder at the same directory level
    example_dir = "../example_data/for_drug_response/"
    drug_info_file = os.path.join(example_dir, "drug_response_14_datasets.csv")
    
    dataset_name = "GSE104987_MCF7"
    method_name = "Example_of_Method"  
    
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    print(f"Starting drug response prediction evaluation using device: {device}")

    # Initialize the evaluator
    evaluator = DrugResponseEvaluator(
        base_dir=example_dir,      
        embed_dir=example_dir,     
        drug_info_path=drug_info_file,
        device=device,
        epochs=20,                 # Use a smaller number of epochs for demonstration/testing
        lr=1e-3,
        patience=5,
        batch_size=64
    )

    # Run 5-fold cross-validation
    result_df = evaluator.run_5_fold(dataset_name=dataset_name, method_name=method_name)
    
    print("\n✅ Five-fold cross-validation test completed successfully! Results:")
    print(result_df)
    
    # Compute the mean of each evaluation metric
    summary = result_df[["MacroF1", "AUROC", "AUPRC"]].mean()
    print("\n📊 Overall average performance:")
    print(summary)
```
