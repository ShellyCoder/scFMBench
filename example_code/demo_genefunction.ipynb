```python
import os
import torch
import pandas as pd
from GeneFunctionEvaluator import GeneFunctionEvaluator
```

```python
if __name__ == "__main__":
    # Paths pointing to the directory
    data_dir = "../example_data/for_gene_function/"
    emb_file = os.path.join(data_dir, "aligned_embeddings.pkl")
    tissue_file = os.path.join(data_dir, "HPA_tissue_specific_genes.pkl")

    # 1. Instantiate the evaluator
    evaluator = GeneFunctionEvaluator(
        emb_file_path=emb_file,
        tissue_genes_path=tissue_file
    )

    # 2. Specifically extract GeneCompass data for testing
    target_model = "GeneCompass"
    
    # Execute 5-fold calculation and return a DataFrame
    df_results = evaluator.run_5_fold(model_name=target_model)

    print("\n========== 📊 Final Results ==========")
    print(df_results)
    
    print("\n========== 📈 Average Metrics ==========")
    print(df_results[["Accuracy", "MacroF1", "AUC", "AUPR"]].mean())
```
